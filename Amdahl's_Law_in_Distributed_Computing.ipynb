```python
#######################################
# Amdahl's Law in Distributed Computing
#######################################

# Amdahl's Law is an equation in computer architecture that states the 
# possible speedup in latency of task execution at a certain workload 
# that may be expected of a system with greater resources.
```

```python
# Initializing variables 
T_1 = 100
S = 0.01
P = 0.99
```

```python
# Amdahl's argument
def Amdahl(n):
    return S*T_1 + (P*T_1)/n
```

```python
Amdahl(100)
```

```python
# Graphing Amdahl argument 
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1, 100000)
plt.plot(x,Amdahl(x), color='red')

plt.show()
```
